# Applied Statistics and Inference Assignment

Question and Answer Format

## Question 1

**What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making?**

**Answer:** Type I Error occurs when a true null hypothesis is rejected. Type II Error occurs when a false null hypothesis is not rejected. Type I errors may cause unnecessary actions, while Type II errors may result in missed opportunities or risks.

## Question 2

**What is the P-value in hypothesis testing, and how should it be interpreted?**

**Answer:** The P-value measures the probability of obtaining the observed result if the null hypothesis is true. A small P-value (typically less than 0.05) indicates strong evidence against the null hypothesis.

## Question 3

**Explain the difference between a Z-test and a T-test.**

**Answer:** A Z-test is used when the population standard deviation is known and the sample size is large. A T-test is used when the population standard deviation is unknown, especially for smaller samples.

## Question 4

**What is a confidence interval?**

**Answer:** A confidence interval is a range of values likely to contain the true population parameter. A larger margin of error produces a wider interval, while a smaller margin of error produces a narrower interval.

## Question 5

**Describe the purpose and assumptions of ANOVA.**

**Answer:** ANOVA compares the means of three or more groups. Assumptions include independence, normality, and equal variances across groups.
